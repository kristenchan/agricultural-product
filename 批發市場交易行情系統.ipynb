```python
import pandas as pd
import json
import requests
import datetime
from pyquery import PyQuery
import numpy as np 
from random import choice
import time, re
```

```python
def create_assist_date(datestart = None,dateend = None):
    if datestart is None:
        datestart = '2019.01.01'
    if dateend is None:
        dateend = datetime.datetime.now().strftime('%Y.%m.%d')
    datestart=datetime.datetime.strptime(datestart,'%Y.%m.%d')
    dateend=datetime.datetime.strptime(dateend,'%Y.%m.%d')
    date_list = []
    date_list.append(str(int(datestart.strftime('%Y'))-1911)+datestart.strftime('.%m.%d'))
    while datestart<dateend:       
        datestart+=datetime.timedelta(days=+1)
        date_list.append(str(int(datestart.strftime('%Y'))-1911)+datestart.strftime('.%m.%d'))
    return date_list
```

### 農產品交易行情
#### https://data.coa.gov.tw/Query/ServiceDetail.aspx?id=037
##### 無法下載 100年以前的數據

```python
dataframe_output_list = []
dates =  create_assist_date(datestart = "2020.04.20")

for date in dates:
    print(date)
    #payload = {'StartDate': date, 'EndDate': date} 
    payload = {'StartDate': date, 'EndDate': date} 
    response= requests.get('https://data.coa.gov.tw/Service/OpenData/FromM/FarmTransData.aspx?$top=50000&$skip=0', params=payload)
    result = json.loads(response.text)
    if result:
        dataframe_output_list.append(pd.DataFrame(result))
        
final_dataframe_output = pd.concat(dataframe_output_list, axis=0, ignore_index=True)
final_dataframe_output['年']=final_dataframe_output['交易日期'].str[0:3] #整理交易日期,整理表格
final_dataframe_output['月']=final_dataframe_output['交易日期'].str[4:6]
final_dataframe_output['日']=final_dataframe_output['交易日期'].str[-2:]
final_dataframe_output["年"]=pd.to_numeric(final_dataframe_output["年"], errors = 'coerce')+1911
final_dataframe_output["年"]=final_dataframe_output["年"].astype(str).str[0:4]
final_dataframe_output["Date"]=final_dataframe_output["年"]+"-"+final_dataframe_output["月"]+"-"+final_dataframe_output["日"]

final_dataframe_output.to_csv('農產品交易行情.csv')
```

```python
def crawler(date_input):
    date = date_input.split(".")[0].zfill(3) + "/" + date_input.split(".")[1] + "/" + date_input.split(".")[2]
    session = requests.Session()    
    url_homepage = "https://amis.afa.gov.tw/veg/VegProdDayTransInfo.aspx"
    res = session.get(url_homepage)
    time.sleep(0.5) 
    ASP_NET_SessionId = res.cookies.values()[0]
    VIEWSTATEGENERATOR =re.findall('<input type="hidden" name="__VIEWSTATEGENERATOR" id="__VIEWSTATEGENERATOR" value="(.*?)" />', str(res.text))
    EVENTVALIDATION =re.findall('<input type="hidden" name="__EVENTVALIDATION" id="__EVENTVALIDATION" value="(.*?)" />', str(res.text))
    VIEWSTATE =re.findall('<input type="hidden" name="__VIEWSTATE" id="__VIEWSTATE" value="(.*?)" />', str(res.text)) 
    header = header_data_fn(user_agents = user_agents_fn(), 
                            SessionId = ASP_NET_SessionId)
    bd = body_fn(VIEWSTATE, VIEWSTATEGENERATOR, EVENTVALIDATION, 
                 date = date)
    res = session.post(url_homepage, headers = header, data = bd) 
    my_list = PyQuery(res.text)(".main_main+ .main_main_02 td , .main_main td").map(lambda i,e:PyQuery(e).text())
    if my_list == []:
        my_list = pd.DataFrame(np.full((1,9), "-"),
                               columns = ["市場","產品","上價","中價","下價","平均價","跟前一交易日比","交易量","跟前一交易日比"])
    else:
        my_list = pd.DataFrame(np.array(my_list).reshape(int(len(my_list)/9), 9), 
                               columns = ["市場","產品","上價","中價","下價","平均價","跟前一交易日比","交易量","跟前一交易日比"])
    my_list['交易日期'] = date_input
    print(date_input)
    time.sleep(0.5) 
    return(my_list)
```

```python
date_list = create_assist_date(datestart = '1996.01.01',dateend = "2011.12.31")
result = []
for num in date_list:
    result.append(crawler(num))    
result = pd.concat(result, axis=0, ignore_index=True)
result
```
